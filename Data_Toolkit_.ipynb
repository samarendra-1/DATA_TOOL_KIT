 Explain the following terms

 The difference in np.array, np.asarray and np.asanyarrayX

  The difference between Deep copy and shallow copy

np.array()

Purpose: Creates a new array from an object, regardless of whether the object is already an array or not.

Behavior: It always creates a copy of the input data, unless the copy parameter is set to False and the input is already an array of the desired type.

Usage: Use this function when you want to ensure that a new array is created.

```python
import numpy as np

# Creating a new array from a list
a = np.array([1, 2, 3])
print(a)  # Output: [1 2 3]

# Creating a new array from another array
b = np.array(a)
print(b)  # Output: [1 2 3]
print(b is a)  # Output: False (different objects)
```

np.asarray()

Purpose: Converts an input to an array, but does not create a copy if the input is already an array.

Behavior: It returns the input itself if it is already an array of the desired type. Otherwise, it creates a new array.

Usage: Use this function to convert inputs to arrays without unnecessary copying, which can improve performance when dealing with large datasets.

```python
import numpy as np

# Converting a list to an array
a = np.asarray([1, 2, 3])
print(a)  # Output: [1 2 3]

# Converting an array (no copy is made)
b = np.asarray(a)
print(b)  # Output: [1 2 3]
print(b is a)  # Output: True (same object)
```

np.asanyarray()

Purpose: Converts an input to an array, but if the input is a subclass of np.ndarray, it returns the input as is.

Behavior: Similar to np.asarray(), but it preserves subclasses of np.ndarray (e.g., np.matrix or custom array subclasses).

Usage: Use this function when working with array subclasses and you want to preserve the subclass type.

```python
import numpy as np

# Converting a list to an array
a = np.asanyarray([1, 2, 3])
print(a)  # Output: [1 2 3]

# Converting an array (no copy is made)
b = np.asanyarray(a)
print(b)  # Output: [1 2 3]
print(b is a)  # Output: True (same object)

# Converting a matrix (preserves the matrix type)
m = np.matrix([[1, 2], [3, 4]])
n = np.asanyarray(m)
print(n)
print(type(n))

```

Shallow Copy

A shallow copy of an object is a new object whose elements are references to the same elements found in the original object. This means that the copied object itself is new, but the elements within it are not copied; they are the same as those in the original object. In other words, a shallow copy only copies the outermost container, not the nested objects within.

Deep Copy

A deep copy of an object is a new object with recursively copied elements from the original object. This means that both the outer container and the nested objects within are copied. As a result, modifying the deep copy does not affect the original object and vice versa.



Level of Copying:

Shallow Copy: Copies only the top-level structure, not the nested objects.

Deep Copy: Recursively copies all objects, including nested objects.

Independence of Copies:

Shallow Copy: Changes to nested objects in the copy will affect the original object since they share references.

Deep Copy: Changes to nested objects in the copy do not affect the original object as they are fully independent.

Performance:

Shallow Copy: Typically faster and uses less memory since it only copies references.

Deep Copy: Slower and uses more memory because it creates new copies of all nested objects.

```python
# shallow copy

import copy

# Original list
original_list = [1, 2, [3, 4]]

# Shallow copy of the original list
shallow_copy_list = copy.copy(original_list)

# Modify the nested list in the shallow copy
shallow_copy_list[2][0] = 'a'

print("Original list:", original_list)
print("Shallow copy:", shallow_copy_list)
```

```python
# deeP COPY

import copy

# Original list
original_list = [1, 2, [3, 4]]

# Deep copy of the original list
deep_copy_list = copy.deepcopy(original_list)

# Modify the nested list in the deep copy
deep_copy_list[2][0] = 'a'

print("Original list:", original_list)
print("Deep copy:", deep_copy_list)
```

Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:                                


 a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic time-series data
np.random.seed(42)
dates = pd.date_range(start='2024-01-01', periods=1000, freq='D')
temperature = np.random.normal(loc=25, scale=5, size=1000)
humidity = np.random.normal(loc=60, scale=10, size=1000)

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plotting Temperature and Humidity
plt.figure(figsize=(10, 6))

# Left y-axis for Temperature
plt.plot(df['Date'], df['Temperature'], color='b', label='Temperature')
plt.ylabel('Temperature (°C)')

# Right y-axis for Humidity
ax2 = plt.twinx()
ax2.plot(df['Date'], df['Humidity'], color='r', label='Humidity')
ax2.set_ylabel('Humidity (%)')

# X-axis label and title
plt.xlabel('Date')
plt.title('Temperature and Humidity Over Time')

# Show both legends
lines, labels = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```

d) Set the title of the plot as 'Histogram with PDF Overlay'



```python
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black', label='Data')

# Overlay the PDF (probability density function)
x_values = np.linspace(min(data), max(data), 1000)
pdf_values = (1 / (np.sqrt(2 * np.pi) * 1)) * np.exp(-(x_values - 0)**2 / (2 * 1**2))
plt.plot(x_values, pdf_values, color='red', label='PDF')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

plt.show()
```

Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random arrays
np.random.seed(42)
x = np.random.uniform(low=-10, high=10, size=1000)
y = np.random.uniform(low=-10, high=10, size=1000)

# Determine quadrant for each point
quadrant = np.zeros(1000)
quadrant[(x > 0) & (y > 0)] = 1
quadrant[(x < 0) & (y > 0)] = 2
quadrant[(x < 0) & (y < 0)] = 3
quadrant[(x > 0) & (y < 0)] = 4

# Create a DataFrame for scatter plot
scatter_df = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': quadrant})

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', hue='Quadrant', data=scatter_df, palette='Set1', legend='full')
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(title='Quadrant')

plt.show()

```

With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x values (from 0 to 2*pi)
x = np.linspace(0, 2*np.pi, 1000)

# Compute the sine function
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)')

# Add the sine wave line
p.line(x, y, line_width=2, legend_label='sin(x)', color='blue')

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
output_notebook()
show(p)
```

Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'

```python
pip install bokeh
```

```python
import random
from bokeh.io import show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

# Example data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(10, 100) for _ in categories]

# Create a color map based on values
colors = factor_cmap('categories', palette=Viridis256, factors=categories)

# Prepare the data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create the figure
p = figure(x_range=categories, min_height=400, min_width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories", line_color='white', fill_color=colors)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Remove the grid
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```
